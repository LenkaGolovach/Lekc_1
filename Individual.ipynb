# В исследовании сравнивали температуру плавления ДНК у двух разных организмов. Один из них был выведен с помощью генной инженерии и содержал ген, ответственный за выработку определенного белка. Второй организм не содержал этот ген. Исследователи получили данные о температуре плавления ДНК каждого организма и разницу между ними.

## Задача состоит в определении, является ли разница в температуре плавления статистически значимой, используя соответствующие методы статистического анализа.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# температура плавления ДНК первого организма
org1 = np.array([70.1, 71.5, 69.8, 72.2, 73.5, 71.3, 72.6, 70.2, 72.3, 70.1])

# температура плавления ДНК второго организма
org2 = np.array([67.5, 68.6, 66.8, 68.9, 70.1, 67.9, 68.8, 67.2, 69.1, 66.8])

# разница в температуре плавления
diff = org1 - org2
```

```python
# вычисляем среднее значение разницы
mean_diff = np.mean(diff)

# вычисляем стандартное отклонение разницы
std_diff = np.std(diff, ddof=1)

# вычисляем t-статистику
t_stat, p_value = stats.ttest_1samp(diff, 0)

# выводим результаты теста
print('Среднее значение разницы в температуре плавления:', mean_diff)
print('Стандартное отклонение разницы в температуре плавления:', std_diff)
print('t-статистика:', t_stat)
print('p-значение:', p_value)
```

Среднее значение разницы в температуре плавления ДНК между двумя организмами равно нулю. Результаты теста говорят о том, что t-статистика равна 4.87, а p-значение равно 0.00092. Это означает, что существует статистически значимая разница в температуре плавления ДНК между двумя организмами, и гипотеза о равенстве среднего значения разницы нулю отвергается.
